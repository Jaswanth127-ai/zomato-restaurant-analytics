```python
final_df.columns
```

```python
final_df.groupby(by=['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(by=['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

observation: 
1. when rating is between 4.5 to 4.9 is excellent
2. when rating is between 4.0 to 4.4 is very good
3. when rating is between 3.5 to 3.9 is good
4. when rating is between 2.5 to 3.4 is average
5. when rating is between 1.8 to 2.4 is poor
6. when rating is 0.0 is not rated

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['grey','red','orange','yellow','green','green'])
```

observation :
1. not rated count is very high
2. maximum no of ratings are from 2.7 to 4.2

```python
sns.countplot(x='Rating color',data=ratings)
```
